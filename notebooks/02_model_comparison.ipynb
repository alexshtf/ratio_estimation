# Model Comparison Demo

A compact comparison between the proximal ratio learner and a simple linear baseline.

```python
import numpy as np
import pandas as pd

from experiments.baselines import LinearRegressionBaseline
from experiments.data import generate_dataset
from experiments.evaluate import run_panel
from ratio_estimation.models import RatioProximalLearner, SoftplusLink
```

```python
dataset = generate_dataset(n_groups=25, history_length=6, rng=np.random.default_rng(0))

scores = {
    "proximal-softplus": run_panel(
        dataset,
        model_factory=lambda: RatioProximalLearner(
            link=SoftplusLink(),
            step_size=0.1,
            regularization=1.0,
        ),
    ),
    "linear-baseline": run_panel(
        dataset,
        model_factory=lambda: LinearRegressionBaseline(
            dimension=6,
            step_size=0.1,
            regularization=1.0,
            inverse=False,
        ),
    ),
}

pd.Series(scores, name="weighted mean log error").sort_values()
```
