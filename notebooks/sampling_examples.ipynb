```python
import simulation as sim
import numpy as np
import matplotlib.pyplot as plt
```

```python
hours = np.arange(24 * 14)
raw_trend = sim.periodic_trend(hours) + sim.global_trend(hours)
```

```python
trend = np.exp(raw_trend)
plt.plot(trend, label='true_trend')
plt.plot(hours, sim.sample_poisson(trend), label='samples')
plt.legend()
plt.show()
```

```python
trend = 5 * np.logaddexp(0, raw_trend) ** 2
plt.plot(trend, label='true_trend')
plt.plot(hours, sim.sample_nb_with_dispersion(10 * trend) / 10, label='samples')
plt.legend()
plt.show()
```

```python
hours, true_trend, hourly_spend, hourly_count = sim.sample_campaign()
plt.plot(hours, true_trend, label='True trend')
plt.plot(hours, hourly_spend, label='Spend')
plt.plot(hours, hourly_count, label='Count')
plt.show()
```
