# Sampling Demo

A small end-to-end example that samples one synthetic campaign and builds autoregressive features.

```python
import matplotlib.pyplot as plt
import numpy as np

from ratio_estimation.features import AutoregressiveRatioFeatures
from ratio_estimation.simulation import sample_campaign
```

```python
campaign = sample_campaign(rng=np.random.default_rng(0))
feature_builder = AutoregressiveRatioFeatures(history_length=12)

feature_snapshots = []
for spend, count in zip(campaign.spend, campaign.count, strict=True):
    feature_snapshots.append(feature_builder.features().copy())
    feature_builder.update(spend, count)

feature_snapshots = np.asarray(feature_snapshots)
campaign.hours[:5], campaign.true_ratio[:5], feature_snapshots[:2, :4]
```

```python
plt.figure(figsize=(10, 4))
plt.plot(campaign.hours, campaign.true_ratio, label="true ratio")
plt.plot(campaign.hours, campaign.spend, label="spend", alpha=0.7)
plt.plot(campaign.hours, campaign.count, label="count", alpha=0.7)
plt.legend()
plt.tight_layout()
```
